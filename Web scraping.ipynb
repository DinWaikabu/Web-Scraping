```python
import requests
```

```python
r  = requests.get("https://www.transtv.co.id/schedule/program")
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
results = soup.find_all('div', attrs={'class':'grid-sc'})
```

## Build Data Set

```python
raw_data = []

for result in results:
    jam = result.find('div',  attrs={'class':'bubu grid-sc-hour'}).text
    nama_program = result.find('div', attrs={'class':'bubu grid-sc-name'}).text
    title = result.find('div',  attrs={'class':'bubu grid-sc-title'}).text
    raw_data.append((jam, nama_program, title))
```

```python
import pandas as pd

df = pd.DataFrame(raw_data, columns=['jam', 'nama_program', 'title'])
print(df)
```

```python
#save to csv

df.to_csv('jadwal_trans7.csv', index=False)
```
